```python
import requests
from bs4 import BeautifulSoup

# Liste des URLs à scraper
urls = [
    "https://www.coachesvoice.com/cv/4-3-3-formation-klopp-arteta-xavi/",
    "https://www.coachesvoice.com/cv/4-5-1-formation-football-tactics/",
    "https://www.coachesvoice.com/cv/4-3-3-formation-klopp-arteta-xavi/",
    "https://www.coachesvoice.com/cv/3-5-2-formation-conte-mourinho-guardiola/",
    "https://www.coachesvoice.com/cv/overloads-football-tactics-explained-guardiola-liverpool-messi/",
    "https://www.coachesvoice.com/cv/what-is-a-box-midfield-football-tactics-liverpool/",
    "https://www.coachesvoice.com/cv/rotations-football-tactics-explained-klopp-liverpool-bielsa-leeds/",
    "https://www.coachesvoice.com/cv/4-4-2-diamond-football-tactics-explained-klopp-potter-allegri/",
    "https://www.coachesvoice.com/cv/switches-of-play-football-tactics-explained/",
    "https://www.coachesvoice.com/cv/inverted-full-backs-guardiola-cancelo-trent-alexander-arnold-lahm-football-tactics/",
    "https://www.coachesvoice.com/cv/half-spaces-football-tactics-explained/",
    "https://www.coachesvoice.com/cv/zonal-marking-football-tactics-explained-mourinho-conte-simeone/",
    "https://www.coachesvoice.com/cv/number-8-football-tactics-henderson-kroos-modric-iniesta/",
    "https://www.coachesvoice.com/cv/wing-backs-football-tactics-explained-conte-tuchel/",
    "https://www.coachesvoice.com/cv/modern-goalkeeper-football-tactics-explained-alisson-ederson-ter-stegen/",
    "https://www.coachesvoice.com/cv/defensive-transitions-football-tactics-explained-chelsea-barcelona/",
    "https://www.coachesvoice.com/cv/low-block-football-tactics-explained-simeone-dyche-mourinho/",
    "https://www.coachesvoice.com/cv/4-4-2-football-tactics-ferguson-simeone-hasenhuttl-dyche/",
    "https://www.coachesvoice.com/cv/counter-pressing-gegenpressing-football-tactics-explained-klopp-guardiola-bielsa-hasenhuttl/",
    "https://www.coachesvoice.com/cv/mid-block-football-tactics-explained/",
    "https://www.coachesvoice.com/cv/attacking-transitions-football-tactics-explained-klopp-liverpool-mourinho/"


    # Ajoute d'autres URLs ici...
]

# Créer une liste pour stocker les données
dataset = []

# Loop à travers chaque URL
for url in urls:
    # Faire une requête GET pour obtenir le contenu de la page
    response = requests.get(url)
    content = response.content

    # Analyser le contenu HTML
    soup = BeautifulSoup(content, "html.parser")

    # Trouver tous les éléments <h2> avec la classe "wp-block-heading"
    titles = soup.find_all("h2", class_="wp-block-heading")

    # Parcourir les titres et trouver le texte associé à chaque titre
    for title in titles:
        # Récupérer le titre
        title_text = title.get_text(strip=True)

        # Trouver le texte associé en utilisant la méthode find_next_sibling
        next_sibling = title.find_next_sibling()
        if next_sibling.name == 'p':
            text_content = next_sibling.get_text(strip=True)
        else:
            text_content = ""

        # Ajouter le titre et le texte associé à la dataset
        dataset.append((title_text, text_content))

# Afficher la dataset
for data in dataset:
    print(data)
```
